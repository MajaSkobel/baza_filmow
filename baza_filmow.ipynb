```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_movies = pd.read_csv('tmdb_movies.csv')
df_genres = pd.read_csv('tmdb_genres.csv')
df_movies.head()
```

```python
df_genres.head()
```

```python
#Zadanie 1
q3=df_movies['vote_count'].quantile(0.75)
q3
```

```python
df_movies[df_movies['vote_count']>q3].sort_values(by='vote_average',ascending=False).head(10)
```

```python
#Zadanie 2

df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])

def select_10to16(x):
    if (x> pd.to_datetime('2009-12-31')) & (x<pd.to_datetime('2017-01-01')):
        return True
    else:
        return False
        
df_movies_10to16=df_movies[df_movies['release_date'].apply(lambda x: select_10to16(x))].copy()
df_movies_10to16.head()
```

```python
df_movies_10to16['release_year']=df_movies_10to16['release_date'].dt.year
df_movies_10to16
```

```python
plot_10to16=df_movies_10to16.pivot_table(values=['revenue','budget'],index='release_year',aggfunc='mean')
plot_10to16.reset_index(inplace=True)
plot_10to16
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(plot_10to16['release_year'],plot_10to16['budget'],c='red',lw=3,label='budget')
axes.bar(plot_10to16['release_year'],plot_10to16['revenue'],label='revenue')
axes.legend(loc=(1.05,0.9))
axes.set_title('Średni budżet i przychód filmów w latach 2010-2016')

def million(x, pos):
        return '{:2.1f}M'.format(x*1e-6)

formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)
```

```python
#Zadanie 3
df_genres.rename(columns={'Unnamed: 0':'genre_id'},inplace=True)
df_movies=pd.merge(df_movies,df_genres,on='genre_id')
df_movies.head()
```

```python
#Zadanie 4
df_movies['genres'].value_counts().head(1)
#Odpowiedź to drama z 1207 wystąpieniami
```

```python
#Zadanie 5
df_movies.pivot_table(values='runtime',index='genres',aggfunc='mean').sort_values(by='runtime',ascending=False).head()
#Odpowiedź to history
```

```python
#Zadanie 6
movies_history=df_movies[df_movies['genres']=='History']
movies_history
```

```python
movies_history['runtime'].hist(bins=10)
plt.xlabel('Runtime in minutes')
plt.ylabel('Number of appearances')

from matplotlib.ticker import MaxNLocator

plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
```
